## 1. Introduction: What is RAG-KG?

**RAG-KG** stands for Retrieval-Augmented Generation with Knowledge Graphs. It is a customer service question-answering system that combines:
- A knowledge graph (Neo4j) built from historical support tickets
- A vector database (Qdrant) for semantic search
- Local Large Language Models (LLMs) via Ollama for answer generation
- A FastAPI server for RESTful queries

**Key Features:**
- Works fully offline (no cloud dependencies)
- Optimized for 16GB RAM (scalable to 32GB+)
- Supports advanced coding/research LLMs
- Easy migration and backup

# 15. Common Errors & Solutions

### Docker
- **Error:** `failed to connect to the docker API ... The system cannot find the file specified.`
  - **Solution:** Ensure Docker Desktop is running. Run `docker ps` to verify.
- **Error:** `port is already allocated`
  - **Solution:** Change the port in `.env` or Docker Compose, or stop the conflicting service.
- **Error:** `no space left on device`
  - **Solution:** Free up disk space or increase Docker's disk allocation.

### Python & Pip
- **Error:** `pip: The term 'pip' is not recognized ...`
  - **Solution:** Activate your virtual environment or use `python -m pip`.
- **Error:** `ModuleNotFoundError: ...`
  - **Solution:** Run `pip install -r requirements.txt` in the active environment.

### Ollama
- **Error:** `pull model manifest: file does not exist`
  - **Solution:** The model name is incorrect or not available in Ollama's registry. Use `ollama search <keyword>`.
- **Error:** `out of memory` or container crashes
  - **Solution:** Use smaller models or increase system RAM.

### Neo4j
- **Error:** `Unable to connect to Neo4j`
  - **Solution:** Ensure the Neo4j container is running and the port is correct.
- **Error:** `Authentication failed`
  - **Solution:** Check credentials in `.env` and Neo4j settings.

### Qdrant
- **Error:** `Qdrant connection failed`
  - **Solution:** Ensure Qdrant is running and the port matches your config.

### General
- **Error:** `Address already in use`
  - **Solution:** Change the port or stop the conflicting process.
- **Error:** `Permission denied`
  - **Solution:** Run as administrator or check file/folder permissions.

# 16. Tricky Interview Questions (with Answers)

### Beginner
- **Q:** What is a knowledge graph and why is it useful in customer service QA?
  - **A:** A knowledge graph is a structured representation of entities and their relationships. In customer service QA, it enables context-aware, explainable answers by connecting related tickets, products, and issues.
- **Q:** What is the difference between a vector database and a graph database?
  - **A:** A vector database (like Qdrant) stores high-dimensional embeddings for semantic search, while a graph database (like Neo4j) stores entities and relationships for logical traversal.

### Intermediate
- **Q:** How does Retrieval-Augmented Generation (RAG) improve answer quality?
  - **A:** RAG combines retrieval (fetching relevant documents or graph nodes) with generation (LLM-based synthesis), grounding answers in real data and reducing hallucinations.
- **Q:** Why use local LLMs (Ollama) instead of cloud APIs?
  - **A:** Local LLMs provide privacy, offline operation, and cost savings, and can be customized for your domain.
- **Q:** How do you update the system with new data?
  - **A:** Add new raw data, rerun parsing, graph, and embedding scripts, and restart the API server if needed.

### Advanced
- **Q:** How would you scale this system for millions of tickets?
  - **A:** Use distributed Neo4j/Qdrant clusters, batch processing, sharding, and possibly split the API server into microservices.
- **Q:** How can you ensure answer explainability?
  - **A:** Return source nodes/tickets with each answer, and use graph traversal paths to show reasoning.
- **Q:** What are the trade-offs between quantized and full-precision models?
  - **A:** Quantized models use less RAM and are faster, but may lose some accuracy. Full-precision models are more accurate but require more resources.
- **Q:** How would you add support for multi-lingual queries?
  - **A:** Use multi-lingual embedding models and LLMs, and ensure your data pipeline supports Unicode and language detection.
