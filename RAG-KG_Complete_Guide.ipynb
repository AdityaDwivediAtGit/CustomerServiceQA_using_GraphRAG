## 1. Introduction: What is RAG-KG?

**RAG-KG** stands for Retrieval-Augmented Generation with Knowledge Graphs. It is a customer service question-answering system that combines:
- A knowledge graph (Neo4j) built from historical support tickets
- A vector database (Qdrant) for semantic search
- Local Large Language Models (LLMs) via Ollama for answer generation
- A FastAPI server for RESTful queries

**Key Features:**
- Works fully offline (no cloud dependencies)
- Optimized for 16GB RAM (scalable to 32GB+)
- Supports advanced coding/research LLMs
- Easy migration and backup

## 11. Troubleshooting & Optimization

### Common Issues
- Docker not running: Start Docker Desktop
- Port conflicts: Change ports in `.env` and Docker Compose
- Out of memory: Use smaller models or increase RAM
- Model not found: Check model name in Ollama registry

### Performance Tuning
- Use quantized models for lower RAM
- Batch data processing
- Monitor logs for errors

## 13. FAQ & Resources

### FAQ
- **Q:** Can I use my own data?  
  **A:** Yes! Place it in `data/raw/` and rerun the data prep scripts.
- **Q:** How do I add new models?  
  **A:** Use `docker exec ollama ollama pull <model>`.
- **Q:** What if a model fails to pull?  
  **A:** It may not exist in the Ollama registry. Check with `ollama search <keyword>`.
- **Q:** How do I monitor resource usage?  
  **A:** Use Task Manager (Windows) or `htop` (Linux/WSL2).
