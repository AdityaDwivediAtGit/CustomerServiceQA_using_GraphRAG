## 1. Introduction: What is RAG-KG?

**RAG-KG** stands for Retrieval-Augmented Generation with Knowledge Graphs. It is a customer service question-answering system that combines:
- A knowledge graph (Neo4j) built from historical support tickets
- A vector database (Qdrant) for semantic search
- Local Large Language Models (LLMs) via Ollama for answer generation
- A FastAPI server for RESTful queries

**Key Features:**
- Works fully offline (no cloud dependencies)
- Optimized for 16GB RAM (scalable to 32GB+)
- Supports advanced coding/research LLMs
- Easy migration and backup

# 15. Common Errors & Solutions

### Neo4j
- **Error:** `Unable to connect to Neo4j`

### General

# 16. Tricky Interview Questions (with Answers)

### Beginner
- **Q:** What is a knowledge graph and why is it useful in customer service QA?
  - **A:** A knowledge graph is a structured representation of entities and their relationships. In customer service QA, it enables context-aware, explainable answers by connecting related tickets, products, and issues.
- **Q:** What is the difference between a vector database and a graph database?
  - **A:** A vector database (like Qdrant) stores high-dimensional embeddings for semantic search, while a graph database (like Neo4j) stores entities and relationships for logical traversal.

### Intermediate
- **Q:** How does Retrieval-Augmented Generation (RAG) improve answer quality?
  - **A:** RAG combines retrieval (fetching relevant documents or graph nodes) with generation (LLM-based synthesis), grounding answers in real data and reducing hallucinations.
- **Q:** Why use local LLMs (Ollama) instead of cloud APIs?
  - **A:** Local LLMs provide privacy, offline operation, and cost savings, and can be customized for your domain.
- **Q:** How do you update the system with new data?
  - **A:** Add new raw data, rerun parsing, graph, and embedding scripts, and restart the API server if needed.

### Advanced
- **Q:** How would you scale this system for millions of tickets?
  - **A:** Use distributed Neo4j/Qdrant clusters, batch processing, sharding, and possibly split the API server into microservices.
- **Q:** How can you ensure answer explainability?
  - **A:** Return source nodes/tickets with each answer, and use graph traversal paths to show reasoning.
- **Q:** What are the trade-offs between quantized and full-precision models?
  - **A:** Quantized models use less RAM and are faster, but may lose some accuracy. Full-precision models are more accurate but require more resources.
- **Q:** How would you add support for multi-lingual queries?
  - **A:** Use multi-lingual embedding models and LLMs, and ensure your data pipeline supports Unicode and language detection.

## Qdrant: Vector Database Examples

### Connect to Qdrant and List Collections


````python
from qdrant_client import QdrantClient

# Connect to Qdrant
client = QdrantClient(host="localhost", port=6333)

# List all collections
collections = client.get_collections()
print("Collections:", [c.name for c in collections.collections])
````

### Insert Vectors into Qdrant


````python
import numpy as np
from qdrant_client.http.models import PointStruct

# Example: create a collection (if not exists)
client.recreate_collection(
    collection_name="demo_collection",
    vectors_config={"size": 3, "distance": "Cosine"}
)

# Insert some vectors
points = [
    PointStruct(id=1, vector=[0.1, 0.2, 0.3], payload={"label": "A"}),
    PointStruct(id=2, vector=[0.2, 0.1, 0.4], payload={"label": "B"}),
]
client.upsert(collection_name="demo_collection", points=points)
print("Inserted points into demo_collection.")
````

### Search for Similar Vectors


````python
# Search for the most similar vector to [0.1, 0.2, 0.3]
search_result = client.search(
    collection_name="demo_collection",
    query_vector=[0.1, 0.2, 0.3],
    limit=1
)
print("Most similar point:", search_result[0].payload)
````

### Example: Running LLM Inference with Ollama

This example demonstrates how to send a prompt to an Ollama model and receive a response using the REST API. Make sure the model you want to use is already pulled and running in Ollama.

````python
import requests

# Replace 'llama3' with your preferred model name
model = "llama3"
prompt = "What is Retrieval-Augmented Generation (RAG) in NLP?"

response = requests.post(
    "http://localhost:11434/api/generate",
    json={"model": model, "prompt": prompt}
)

print("Ollama response:", response.json()["response"])
````

## Neo4j: Graph Database Usage Examples

This section provides practical code examples for connecting to Neo4j, running Cypher queries, and performing advanced graph operations in Python.

````python
from neo4j import GraphDatabase

# Update these with your Neo4j instance details
uri = "bolt://localhost:7687"
user = "neo4j"
password = "your_password"

driver = GraphDatabase.driver(uri, auth=(user, password))

with driver.session() as session:
    result = session.run("MATCH (n) RETURN n LIMIT 5")
    for record in result:
        print(record)
driver.close()
````

### Example: Creating Nodes and Relationships in Neo4j

This example shows how to create nodes and relationships using Cypher queries from Python.

````python
from neo4j import GraphDatabase

uri = "bolt://localhost:7687"
user = "neo4j"
password = "your_password"

driver = GraphDatabase.driver(uri, auth=(user, password))

with driver.session() as session:
    cypher = """
    CREATE (c:Customer {name: 'Alice'})
    CREATE (a:Agent {name: 'Bob'})
    CREATE (c)-[:ASKED]->(q:Question {text: 'How do I reset my password?'})
    CREATE (a)-[:ANSWERED]->(q)
    RETURN c, a, q
    """
    result = session.run(cypher)
    for record in result:
        print(record)
driver.close()
````

### Example: Advanced Cypher Query – Path Traversal

This example demonstrates how to perform a multi-hop traversal to find all questions answered by a specific agent.

````python
from neo4j import GraphDatabase

uri = "bolt://localhost:7687"
user = "neo4j"
password = "your_password"

driver = GraphDatabase.driver(uri, auth=(user, password))

with driver.session() as session:
    cypher = """
    MATCH (a:Agent {name: 'Bob'})-[:ANSWERED]->(q:Question)
    RETURN q.text AS question
    """
    result = session.run(cypher)
    for record in result:
        print("Answered question:", record["question"])
driver.close()
````
